## Расчет метрик и построение графиков

Рассчитаем и отобразим на графиках метрики трёх видов: продуктовые, метрики электронной коммерции, маркетинговые.

### Продуктовые метрики

#### Метрики пользовательской активности DAU, WAU и MAU

Изучим пользовательскую активность.  
Рассчитаем DAU, WAU и MAU.
Чтобы посчитать недельное и месячное использование, выделим в отдельные столбцы год, месяц и неделю, а также полную дату.

```python
visits['session_year'] = visits['start_ts'].dt.year
visits['session_date'] = visits['start_ts'].dt.date
visits['session_week'] = visits['start_ts'].astype('datetime64[W]')
visits['session_month'] = visits['start_ts'].astype('datetime64[M]')
visits.head()
```

Посчитаем метрики DAU, WAU и MAU

```python
dau_total = int(visits.groupby('session_date').agg({'uid':'nunique'}).mean())
wau_total = int(visits.groupby(['session_year', 'session_week']).agg({'uid':'nunique'}).mean())
mau_total = int(visits.groupby(['session_year', 'session_month']).agg({'uid':'nunique'}).mean())
```

```python
print('DAU = ', dau_total)
print('WAU = ', wau_total)
print('MAU = ', mau_total)
```

Отобразим на графиках изменения метрик пользовательской активности во времени. 

```python
def active_users_plot(df, period, ylabel, title):
    df.groupby(period).agg({'uid':'nunique'}).plot(figsize=(13,9))
    plt.xlabel('Период времени')
    plt.ylabel(ylabel)
    plt.title('Динамика ' + title + ' пользовательской активности')
    plt.show()
```

```python
active_users_plot(visits, 'session_date', 'DAU', 'дневной')
```

Видим один яркий всплеск пользовательской активности в конце ноября (возможно, предновогодний), и один значительный спад в конце марта (до 0, возможно, был какой-то технический сбой в работе сервиса в этот день). По графику заметно, что в зимние месяцы активность пользователей выше, чем в летние, что весьма логично.

```python
active_users_plot(visits, 'session_week', 'WAU', 'недельной')
```

График недельной активности показывает схожую картину с графиком дневной пользовательской активности. Стал заметен рост перед резким спадом активности в марте.

```python
active_users_plot(visits, 'session_month', 'MAU', 'месячной')
```

График месячной пользовательской активности менее информативен. Виден пик в ноябре и большая популярность сервиса в зимние месяцы относительно летних.  
Необходимо также посмотреть на динамику затрат по группам и оценить, есть ли закономерности в изменении размер аудитории относительно затрат

**Sticky factor DAU/WAU, DAU/MAU**  
Рассчитаем показатели лояльности аудитории sticky factor (англ. «липкий фактор»), оценим, как часто пользователи обращаются к сервису.

```python
sticky_wau = dau_total / wau_total
sticky_mau = dau_total / mau_total
print('sticky factor DAU/WAU = {:.2%}'.format(sticky_wau))
print('sticky factor DAU/MAU = {:.2%}'.format(sticky_mau))
```

**Вывод**

У пользовательской активности наблюдается сезонность - в зимний период пользователи более активны, чем в летний.  
Требуется дополнительное изучение причин резких изменений в пользовательской активности. 
Можно предположить, что всплеск в ноябре связан с выбором новогодних мероприятий, детских елок, а также выбор билетов на мероприятия в подарки, а может были проведены специальные маркетиновые мероприятия на увеличение активности пользователей.  
Также следует изучить причины роста, а затем резкого спада активности в конце марта.   
Нужно проследить, были ли проведены некие мероприятия, после которых размер аудитории резко менялся, т.е. посмотреть изменение затрат во времени и по группам.  

Рассчитанные показатели лояльности аудитории говорят нам, что регулярность использования сервиса один и тем же пользователем невысока, т.е. аудитория вовлечена слабо, стабильность пользовательской базы низкая. 

#### Среднее количество посещений в день

Определим, сколько раз за день пользователи в среднем заходят на сайт (сколько сессий в день).  

Динамику среднедневного количества уникальных пользователей мы рассмотрели выше (DAU),  
посмотрим, как изменяется во времени среднедневное количество сессий, и сколько сессий в день в среднем приходится на одного уникального пользователя.  

Сгруппируем данные о визитах по дате и посчитаем количество сессий и количество уникальных пользователей в день,  исходя из них узнаем, сколько раз за день пользователи в среднем заходят на сайт

```python
day_sessions = visits.groupby('session_date').agg({'uid':['count', 'nunique']})
day_sessions.columns = ['n_sessions', 'n_users']
day_sessions['sessions_per_user'] = day_sessions['n_sessions'] / day_sessions['n_users']
day_sessions = day_sessions.reset_index()
display(day_sessions.head())
print('В среднем на одного уникального пользователя приходится', day_sessions['sessions_per_user'].mean().round(2), 'посещение в день')
print('Среднее количество сессий в день (количество посещений в день) - ', day_sessions['n_sessions'].mean().round(2))
```

Посмотрим на графиках динамику среднего количества сессий в день (количества посещений в день) и среднего количества сессий на одного уникального пользователя.

```python
def plot_creator(df, parameter_x, parameter_y, x_label, y_label, plt_title):
    plt.figure(figsize=(13,9))
    plt.plot(df[parameter_x], df[parameter_y])
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(plt_title)
    plt.show()
```

```python
plot_creator(day_sessions, 'session_date', 'n_sessions', 'Дата', 
             'Дневное кол-во посещений', 'Динамика количества сессий в день (количества посещений в день)')
```

#### Продолжительность типичной пользовательской сессии  

Найдём среднюю продолжительность сессии (ASL) — посчитаем длину всех сессий и вычислим среднее:

```python
visits['asl'] = (visits['end_ts'] - visits['start_ts']).dt.seconds
visits['asl'].mean()
```

Расчетное время средней сессии 643,5 сек, почти 11 минут. Посмотрим на распределение средней продолжительности сесии:

```python
def distplot_creator(dataframe, parameter, x_label, title_name):
    plt.figure(figsize=(10,5))
    sns.distplot(dataframe[parameter], bins=dataframe[parameter].nunique(), kde=True)
    plt.xlabel(x_label)
    plt.ylabel('Частота')
    plt.title(title_name)
    plt.axvline(x=dataframe[parameter].mean(), linewidth=3, color='b', linestyle=':')
    plt.show()
```

```python
distplot_creator(visits, 'asl', 
                 'Продолжительность сессии, сек.', 'Распределение средней продолжительности сессии')
```

Малоинформативный график, видим длинный хвост из больших значений. Можно предположить, что на рассчитанную выше ASL в значительной мере повлияли выбросы в виде очень больших значений.  
Попробуем убрать выбросы за верхним усом.

```python
def quantile_counter(dataframe, parameter):
    q1 = dataframe[parameter].quantile(0.25)
    q3 = dataframe[parameter].quantile(0.75)
    iqr = q3 - q1
    bottom_lim = q1 - 1.5 * iqr
    up_limit = q3 + 1.5 * iqr
    print('Нижний ус', parameter, ' - ', bottom_lim)
    print('Верхний ус', parameter, ' - ', up_limit)
    print()
    print(dataframe[parameter].describe())
```

```python
quantile_counter(visits, 'asl')
```

```python
distplot_creator(visits.query('asl <= 1920'), 'asl', 
                 'Продолжительность сессии, сек.', 'Распределение средней продолжительности сессии')
```

Распределение скошено влево, полученное среднее значение продолжительности сессии находится справа от наиболее часто встречающихся значений. Рассчитаем моду и медиану.

#### Retention Rate. Cредний Retention Rate на второй месяц «жизни» когорт.

Оценим, сколько пользователей из когорты относительно их изначального числа вернулось в последующие периоды.  

Сформируем когорты:  
Событие - впервые использовали сервис;  
Период - сгруппируем по месяцам первого визита.

```python
first_activity_date = visits.groupby('uid').agg({'session_date':'min'}).reset_index()
first_activity_date.columns = ['uid', 'first_activity_date']
visits = visits.merge(first_activity_date, on='uid') 
```

```python
visits.head(2)
```

```python
visits['first_activity_month'] = visits['first_activity_date'].astype('datetime64[M]')
visits['activity_month'] = visits['session_date'].astype('datetime64[M]')
```

Добавим столбец с месяцем жизни когорты.

```python
visits['cohort_lifetime'] = visits['activity_month'] - visits['first_activity_month']
visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1,'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int')
```

```python
visits.head(2)
```

Сгруппируем данные по когорте и lifetime. Посчитаем для каждой когорты количество активных пользователей на определёный месяц жизни:

```python
cohorts = visits.groupby(['first_activity_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
cohorts.head()
```

Посмотрим, как менялось количество пользователей по когортам

```python
cohort_users_in_lifetime = visits.pivot_table(
    index='first_activity_month', columns = 'cohort_lifetime', 
    values = 'uid', aggfunc = 'nunique')
cohort_users_in_lifetime.fillna('')
```

Заметно, что в зимние месяцы с октября по март исходное количество пользователей в когортах примерно в два раза больше, чем в летние месяцы.   
Более высокая активность в зимние месяцы наблюдалась и при анализе перыдущих метрик. 

Для расчета Retention Rate найдём исходное количество пользователей в когорте (их число на нулевую неделю):

```python
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_activity_month','uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})
cohorts = cohorts.merge(initial_users_count,on='first_activity_month') 
cohorts.head()
```

```python
cohorts['retention_rate'] = cohorts['uid']/cohorts['cohort_users'] 
cohorts.head()
```

Построим сводную таблицу и создадим тепловую карту:

```python
retention_pivot = cohorts.pivot_table(index='first_activity_month',columns='cohort_lifetime',
                                      values='retention_rate',aggfunc='sum') 
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_pivot, annot=True, fmt='.1%', vmax=0.1, linewidths=1, linecolor='white');
```

Найдем средний Retention Rate на второй месяц «жизни» когорт.

```python
print('Cредний Retention Rate на второй месяц «жизни» когорт {:.2%} '.format(retention_pivot[1].mean()))
```

**Вывод**  
Тепловая карта показывает, что в целом клиенты всех когорт возвращаются плохо.  
У новых когорт, начиная с когорты декабря, процент удержания заметно снизился даже на начальных этапах жизни когорты.  
Также видим, что чем старше становится когорта, тем ниже процент удержания пользователей.  К концу рассматриваемого года процент вернувшихся пользователей во всех когортах наиболее низкий. 

Retention Rate у сервиса совсем не высокий. Следует обратить внимание на удержание пользователей, чтобы увеличить выручку.

#### Анализ динамики LTV на покупателя по когортам. Средний LTV по когортам за 6 месяцев. 

Выясним, как меняется LTV на покупателя по когортам. 

```python
orders.head(2)
```

Построим когорты. Сгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа, сложим выручку и посмотрим, какую выручку приносила когорта каждый месяц. 

```python
cohorts = orders.groupby(['first_order_month','order_month']).agg({'revenue': 'sum'}).reset_index()
cohorts.head()
```

Добавим в таблицу cohorts данные о том, сколько людей первый раз совершили покупку в каждый месяц:

```python
orders_cohort_sizes = orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
orders_cohort_sizes.columns = ['first_order_month', 'n_buyers']
orders_cohort_sizes.head()
```

Объединим данные в одну таблицу для расчета `ltv`

```python
ltv_report = pd.merge(orders_cohort_sizes, cohorts, on='first_order_month')
ltv_report.head()
```

LTV нужно считать из валовой прибыли. Умножим выручку на маржинальность. 
И добавим стоблец с возрастом когорты. Рассчитаем LTV.  

Маржинальность сервиса — 100%.

```python
margin_rate = 1
    
ltv_report['gp'] = ltv_report['revenue'] * margin_rate
ltv_report['cohort_lifetime'] = (ltv_report['order_month'] - ltv_report['first_order_month']) / np.timedelta64(1, 'M')
ltv_report['cohort_lifetime'] = ltv_report['cohort_lifetime'].round().astype('int')
ltv_report['ltv'] = ltv_report['gp'] / ltv_report['n_buyers']

ltv_report.head()
```

Составим сводную таблицу по когортам. Посчитаем LTV накопительным итогом, т.к. LTV - это накопительная выручка когорты в пересчёте на количество людей. 

```python
ltv_report_pivot = ltv_report.pivot_table(
        index='first_order_month', 
        columns='cohort_lifetime', 
        values='ltv', 
        aggfunc='sum').cumsum(axis=1).round(3)
    
ltv_report_pivot.round(2).fillna('')
```

Посмотрим тепловую карту.

```python
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohorts: LTV')
sns.heatmap(ltv_report_pivot, annot=True, fmt='.2f', linewidths=1, linecolor='white');
```

Выделяются когорты июня и сентября 2017, покупатели, входящие в них, приносят наибольшее количество прибыли.  
С января 2018 можно отметить общее снижение прибыльности когорт.

Рассчитаем средний LTV по когортам за 6 месяцев

```python
ltv_report_pivot[5].mean()
```

В среднем каждый покупатель за 6 месяцев приносит по 7.969 у.е. валовой прибыли.

**Вывод**  
Сентябрьская когорта выделялась бОльшим средним чеком относительно других когорт.  
А в июньской когорте отмечен наибольший процент удержания клиентов и наибольшее количество покупок на одного клиента за рассматриваемый год.  
Вероятно, за счет этого клиенты из данных когорт принесли больше прибыли, чем остальные.  
Когорты, начиная с января 2018, являются менее прибыльными в сравнении с другими. Следует поискать варианты увеличения доходности из этих когорт.

#### ROMI по когортам в разрезе источников. Сравнение окупаемости за одинаковые периоды жизни когорт.   

Изучим, насколько окупились расходы

```python
orders.head(2)
```

Найдем количество покупателей в когортах в разрезе источников

```python
cohort_source_sizes = orders.groupby(['first_order_month', 'source_id']).agg({'uid': 'nunique'}).reset_index()
cohort_source_sizes.columns = ['first_order_month', 'source_id', 'n_buyers']
cohort_source_sizes
```

Рассчитаем доход по когортам в каждый месяц в разрезе источников

```python
cohorts_source_revenue = orders.groupby(['first_order_month','order_month', 'source_id']).agg({'revenue':'sum'}).reset_index()
cohorts_source_revenue
```

Рассчитаем расходы в каждый месяц в разрезе источников

```python
costs.head()
```

```python
costs_source_month = costs.groupby(['costs_month', 'source_id']).agg({'costs':'sum'}).reset_index()
costs_source_month
```

Объединим таблицы с количеством покупателей, доходами и расходами по когортам в каждый месяц в разрезе источников

```python
cohorts_source_report = pd.merge(cohort_source_sizes, cohorts_source_revenue,
                                 on=['first_order_month', 'source_id'])

cohorts_source_report = pd.merge(cohorts_source_report, costs_source_month,
                                 left_on=['first_order_month', 'source_id'],
                                 right_on=['costs_month', 'source_id'])

cohorts_source_report
```

LTV нужно считать из валовой прибыли. Так как по условиям маржинальность сервиса — 100%, при расчете LTV продолжим пользоваться выручкой `revenue`

Выделим возраст когорт

```python
cohorts_source_report['age'] = ((cohorts_source_report['order_month'] - cohorts_source_report['first_order_month']) 
                                / np.timedelta64(1, 'M')).round().astype('int')
cohorts_source_report
```

Рассчитаем LTV, CAC и ROMI

```python
cohorts_source_report['ltv'] = cohorts_source_report['revenue'] / cohorts_source_report['n_buyers']
cohorts_source_report['cac'] = cohorts_source_report['costs'] / cohorts_source_report['n_buyers']
cohorts_source_report['romi'] = cohorts_source_report['ltv'] / cohorts_source_report['cac']
cohorts_source_report
```

Создадим общий отчет по ROMI по когортам в разрезе источников

```python
cohorts_source_report_romi = cohorts_source_report.pivot_table(
    index=['first_order_month', 'source_id'],
    columns='age',
    values='romi',
    aggfunc='sum'
).cumsum(axis=1)
cohorts_source_report_romi
```

Самый высокий LTV отмечался в когортах июня и сентября 2017, т.е. покупатели, входящие в них, принесли наибольшее количество прибыли.  
Посмотрим как окупались источники привлечения покупателей в этих когортах.

```python
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('ROMI по источникам', fontsize=14)
sns.heatmap(cohorts_source_report_romi.loc['2017-06-01'],
            annot=True, fmt='.2f', linewidths=1, linecolor='white');
```

В когорте июня 2017 лучше всего окупаются вложения в источник привлечения №1.  Источник №3, в который вкладывается наибольшая сумма средств, окупается хуже всего. Источник №2, который тоже в топе затрат, окупается плохо. 

```python
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('ROMI по источникам', fontsize=14)
sns.heatmap(cohorts_source_report_romi.loc['2017-09-01'],
            annot=True, fmt='.2f', linewidths=1, linecolor='white');
```

В когорте сентября 2017 ситуация несколько иная, лучше всего окупаются расходы по источнику №5. Но, как и в когорте июня, можно отметить, что самые бюджетозатратные источники №3 и №2 окупаются плохо.

Рассчитаем средний ROMI на первый месяц жизни по источникам

```python
romi_source_mean = cohorts_source_report_romi.reset_index().groupby(['source_id'])[0].mean().reset_index()
romi_source_mean.columns = ['source_id', 'romi_1m']
romi_source_mean
```

Добавим средний ROMI по всем месяцам жизни когорт

Источник №3, в который вкладывается наибольшее количество средств, имеет самую низкую окупаемость. Стоит подумать о сокращении затрат на этот источник и перераспределении этих срещдств на другие более перспективные источники.   
Источники 2,4,5, имеющие довольно большие доли в структуре затрат, окупаются вяло. 
Стоит попробовать развивать источник №9, его доля в затратах маленькая, а окупаемость выше многих других источников, бюджет которых больше.

### Итоги подсчета метрик  

Рассчитанные метрики показали, что чаще всего пользователи посещают сервис не чаще одного раза в день и тратят на сессию немного времени - до 5 минут. Большинство покупок совершается пользователями в первый день посещения сервиса в течение первого часа, и более того в течение первых нескольких минут. Складывается впечатление, что большинство пользователей приходят в сервис целенаправленно за совершением покупки, т.е. произвести оплату уже заранее выбранного товара. Не проводят время в самом сервисе, просматривая его.  

Регулярность использования сервиса один и тем же пользователем невысока, т.е. аудитория вовлечена слабо. Коэффициент удержания у сервиса довольно низкий, и, что более критично, с течением времени он имеет тенденцию к дальнейшему снижению.   

Также можно отметить, что для пользовательской активности характерна сезонность - больше активности наблюдается в зимний период, нежели в летний. 

Средний чек не стабилен, есть как периоды роста, так и периоды падения. Т.е. есть возможности для наращивания и удержания показателя, над которыми требуется поработать.  
В среднем каждый покупатель за полгода приносит по 7.969 у.е. валовой прибыли.

Важно выяснить, с чем были связаны аномалии - существенный однодневный рост посещений 24.11.2017 и полное отсутствие посещений 30-31.03.2018. Возможно, причины резкого роста посещений можно использовать для дальнейшего развития. И устранить причины, по которым сервис не работал два дня.
